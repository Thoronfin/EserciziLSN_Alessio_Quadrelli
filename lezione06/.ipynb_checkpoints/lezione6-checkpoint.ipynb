## 6.1

Nel modello di Ising monodimensionale il sistema è formato da un reticolo monodimensionale di spin  $\{ s_i \}$, con ($s_i = \pm 1$), intergenti tra primi vicini con una certa costante di accoppiamento $J$ ed interagenti con un campo emagnetico esterno $h$ tramite la seguente Hamiltoniana:

$$H = -J\sum_{i=1}^N s_i s_{i+1}-\frac{h}{2}\sum_{i=1}^N (s_i + s_{i+1}). $$

imponendo le condizioni periodiche al contorno $s_{N+1} = s_{1}$. In questo modo è possibile ricavare l'espressione analitica dell'energia interna per particella $\frac{U}{N}$, del calore specifico $C$, la suscettività magnetica $\chi$ e della magnetizzazione $M$ in funzione della temperatura $T$ da confrontare con i risultati della simulazione ad $h=0$. Inoltre viene calcolcata, e confrontata con i risultati analitici, la magnetizzazione ad $h=0.02$.


Gli $N_{spin}=50$ spin vengono inizializzati casualmente con probabilità P($s_i = -1 $)=P($s_i = 1 $)= $50 \% $, come se il sistema fosse a temparatura infinita in cui gli spin hanno orientazione casuale poichè il sistema rimane disordinato finito a $T=0$ in una dimensione. Partendo da $T=2$, l'equilibrazione richiede solo poche configurazioni poiché il sistema è già disordinato per cui non è necessaria.
Per simulare il modello di Ising all'equilibrio termico vengono campionati i pesi di Boltzamnn al variare della temperatrua $T \in [0.5,2]$ - $p(\{ s_i \}) = \exp(-\beta H)/Z$ (Z è la funzione di partizione)- con l'algoritmo di Metropolis e il campionatore Gibbs: ad ogni passo vengono proposti $N_{spin}$ cambiamenti di un solo spin in una posizione del reticolo casuale.


Il sistema viene lasciato evolvere, da $T=2$ a a $T=0.5$ per passi di $\Delta T =0.05$, per $N=20$ blocchi ognuno da $L=10^4$ nuovi passi in modo che si ottengano stime non correlate delle grandezze termodinamiche. Inoltre ogni volta che la temperatura viene ridotta di $\Delta T=0.05$ in pochi passi Monte Carlo il sistema raggiunge l'equilibrio per cui non è necessaria l'equlibrazione ad ogni cambio di temperatura. Si osserva che al decrescere della temperatura la probabilità di accettazione di un nuovo stato di spin $s_i$ decresce fino a $P(s_i) \approx 10^{-4}$ a $T=0.5$.

Di seguito sono allegati i grafici, realizzati con i valori ottenuti con il data blocking delle grandezze istantanee con $N=20$, in cui sono confrontati la magnetizzazione, l'energia interna, il calore specifico e la suscettività, ottenuti sia con l'algoritmo di Metropolis sia con il campionatore di Gibbs, con i risultati analitici. Tutti i risultati ottenuti risultano compatibili tra loro e con i valori esatti.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

### Calcolo analitico delle grandezze fisiche del modello di Ising 1D con interazione a primi vicini

```python
points=100
T = np.linspace(0.2,3.0,num=points)
M_0= np.zeros(points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )
heat=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)

X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)

h=0.02 #external field
b = 1/T
l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)
```

### Magnetizzazione

```python
plt.figure(figsize=(15,10))

plt.subplot(2,2,1)
x, y, error = np.loadtxt("mag.result.0", usecols=(0,1,2), unpack='true')
plt.errorbar(x,y,yerr=error, label='Metropolis', marker='x', markersize=10, fmt='o')


x1, y1, error1 = np.loadtxt("mag.result.Gibbs", usecols=(0,1,2), unpack='true')
plt.errorbar(x,y,yerr=error, label='Gibbs', markersize=7, fmt='o')

plt.plot(T, M_0, label="Espressione analitica")
plt.legend()

plt.legend()
plt.grid(True)
plt.title('Magnetizzazione a campo nullo')
plt.xlabel('T')
plt.ylabel(r'$ M / \mu_B $')
plt.plot()

plt.subplot(2,2,2)
x2, y2, error2 = np.loadtxt("mag_campo.result.0", usecols=(0,1,2), unpack='true')
plt.errorbar(x2,y2,yerr=error2, label='Metropolis', markersize=7, fmt='o')

x3, y3, error3 = np.loadtxt("mag_campo.result.Gibbs", usecols=(0,1,2), unpack='true')
plt.errorbar(x3,y3,yerr=error3, label='Gibbs', marker='x', markersize=10, fmt='o')

plt.plot(T, M, label="Espressione analitica")
plt.legend()

plt.grid(True)
plt.title('Magnetizzazione a h=0.02')
plt.xlabel('T')
plt.ylabel(r' M / $\mu_B $')
plt.show()
```
