## Esercizio 2.1

L'integrale $I = \int_0^1 \frac{\pi}{2}\cos(\frac{\pi x}{2}) dx$ è stato calcolato tramite due metodi Monte Carlo diversi:

* campionando dalla distribuzione uniforme in $[0,1)$ una successione di punti $x_i$: $I= (1-0) \sum_{i} \frac{\pi}{2}\cos(\frac{\pi x_i}{2})$.


* Importance sampling: poichè $\frac{\pi}{2}\cos(\frac{\pi x}{2}) $ vale $\frac{\pi}{2}$ in $x=0$, $0$ in $x=1$ ed è monotona decrescente è stata scelta come funzione simile da normalizzare e campionare $P(x)= c \frac{\pi}{2} (1-x)$ poichè varia in maniera monotona tra $\frac{\pi}{2}$ e $0$ ed è non negativa in $[0,1]$. Il coefficiente di normalizzazione si ottiene imponendo che l'area sottesa a $P(x)$ sia $1$: $$\int_0^1 c \frac{\pi}{2} (1-x) \mathrm{d}x = c \frac{\pi}{4} =1. $$ Si ricava che $c= \frac{4}{\pi}$ e quindi $P(x)=2(1-x)$. 
 
 Per campionare $P(x)$ è stato utilizzato il metodo dell'inversione della distribuzione cumulata cioè generando $y$ uniformemente in $[0,1)$ ed ottenendo $x$ tramite l'equazione $x= 1+ \sqrt{1-y}$ . L'integranda risulta quindi essere $  \frac{\frac{\pi}{2}\cos(\pi x/2)}{2(1-x)}$.

L'integrale è stato calcolato $M = 10^4$ volte con entrambi i metodi utilizzando $10^4$ punti ciascuno. I risultati sono stati poi analizzati con il metodo del data blocking dividendo gl $M$ integrali in $N=100$ blocchi da $L=100$ dati l'uno. Come mostra il grafico sottostante, in cui è riportato l'integrale a cui è stato sottratto il valore esatto $1$, entrambi i metodi convergono al valore atteso entro l'errore. Inoltre l'errore sulla stima ottenuta con l'importance sampling risulta ridotto di un fattore $3.77$ rispetto a quello ottenuto campionando la distribuzione uniforme.

```python
x, y, error = np.loadtxt("risultati02.1.1.out", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,y-1, yerr=error, label='Integrale calcolato con distribuzione uniforme')
dev1=error[len(error)-1]


x, y, error = np.loadtxt("risultati02.1.2.out", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,y-1, yerr=error, label='Integrale calcolato con importance sampling')
dev2=error[len(error)-1]


plt.xlabel('N')
plt.ylabel('I-1')

plt.legend()
plt.grid(True)
plt.title(r'$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx - 1$')

plt.show()
```

## Esercizio 2.2

Per realizzare il random walk, che inizi nell'origine $\vec{r} = \vec{0}$ , ogni passo viene determinato come segue:

* **Reticolo cubico:** è stato estratto un numero casuale intero $n={0,1,2,3,4,5}$ per determianre in quale direzione principale viene effettuato il passo di ampiezza $a=1$ (se $n=0$ viene effettuato un passo in direzione $x$ con verso positivo, se $n=2$ viene effettuato un passo in direzione $x$ con verso negativo, etc.).
* **Continuo:** sono generati gli angoli $\theta$ e $\phi$ delle coordinate sferiche da cui si ricava il passo in coordinate cartesiane da sommare alla posizione precedente $\vec{r}=a[\sin(\theta)\cos(\phi), \sin(\theta)\sin(\phi), \cos(\theta)]$.

In entrambi i casi è stata valutata la distanza cartesiana dall'origine al variare del numero dei passi $N$, $\vec{r}_N = a \sqrt{x_N^2+y_N^2+z_N^2}$, ed il suo quadrato $|\vec{r}_N | ^2$.

Sono state calcolate $10^4$ realizzazioni di $|\vec{r}_N | ^2$ per ognuno degli $N=1,2,3,...,100$, su cui è stato effettuato il data blocking con $100$  blocchi da $L= 100$ dati ciascuno. I grafici sottostanti riportano i valori medi di $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$, con il realtivo errore, al variare di dei passi $N$, stimati con il data blocking con $100$ blocchi.

I risultati sono stati fittati con un $f(x)=k\sqrt{x}$ con $k=1$ e mostrano un'ottima aderenza con tale curva, segno del fatto che il random walk sia disceto sia continuo ha un comportamento diffusivo poichè la deviazione standard cresce come la radice del numero dei passi effettuati e che effettivamente il passo è $a=1$ poichè nel random walk $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}= a\sqrt{N}$.

```python
x_t=np.linspace(0,100)
y_t=np.sqrt(x_t)


plt.figure(figsize=(16,6))

plt.subplot(1,2,1)
x, y, error = np.loadtxt("risultati02.2.1.out", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,y, yerr=error, label='dati', fmt='o', marker='x', markersize=5, alpha=0.7)
plt.plot(x_t,y_t, label='fit')
plt.xlabel('passi')
plt.ylabel(r'$\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$')
plt.legend()
plt.grid(True)
plt.title(r'Reticolo cubico di passo reticolare $a = 1$')

plt.subplot(1,2,2)
x, y, error = np.loadtxt("risultati02.2.2.out", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,y, yerr=error, label='dati', fmt='o', marker='x', markersize=5, alpha=0.7)
plt.plot(x_t,y_t, label='fit')
plt.xlabel('passi')
plt.ylabel(r'$\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$')
plt.legend()
plt.grid(True)
plt.title(r'Random walk nel continuo con passo $a =1$')

plt.show()
```
