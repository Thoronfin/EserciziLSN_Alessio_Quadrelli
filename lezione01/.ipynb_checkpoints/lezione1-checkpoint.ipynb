## Esercizio 01.1

### 1

Utilizzando il generatore di numeri casuali fornito, è stato calcolato l'integrale $\langle r \rangle  =\int_o^1 r dr$ $N=100$ volte mediando su blocchi di $L=10^3$ numeri generati casualmente in $[0,1)$. Successivamente è stato implementato il metodo del data blocking, cioè sono state calcolate le medie progressive, sommando progressimanete i $100$ blocchi e dividendoli per il numero di blocchi usati, e il relativo errore tramite la deviazione standard delle media progressiva. Il grafico sottostante mostra che all'aumentare del numeri di blocchi utilizzati, $N$, le medie progressive dei blocchi convergono al valore analitico, $0.5$, con un errore progressivamente decrescente come garantito dal teorema del limite centrale poichè l'integrale è calcolato come una somma di variabili sono indipendenti, identicamente distribuite ed estratte da una distribuzione uniforme con varianza finita.

```python
x, y, error = np.loadtxt("risultati01.1.1.out", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,y-0.5, yerr=error)
plt.xlabel('N')
plt.ylabel(r'$<r>$- (1/2)')
plt.title(r'$<r>$')
plt.grid(True)

plt.show()
```

### 2

Utilizzando lo stesso metodo descritto sopra, è stato calcolato $\sigma^2  =\int_o^1 (r- \langle r \rangle)^2  dr$ che è la varianza della distibuzione degli $r$, cioè della distribuzione uniforme dal momento che gli $r$ vengono generati uniformemente.

Nel grafico sottostante è riportata la deviazione standard con il relativo errore stimati con il metodo del data blocking in funzione del numeri di blocchi, $N$, utilizzati. 
Anche in questo caso, poichè l'integrale è una somma di varabili indipendenti, identicamente distribuite ed estratte da una distribuzione con varianza finita, è applicabile il teorema del limite centrale. Si osserva che la deviazione standard converge, entro l'inceretezza associata, a $\frac{1}{12}$ come mostra il grafico.

```python
x, y, error = np.loadtxt("risultati01.1.2.out", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,y-(1/12),yerr=error)
plt.xlabel('#throws')
plt.ylabel('$<(r-0.5)^2>$-1/12')
plt.title('$\sigma^2$')
plt.grid(True)

plt.show()
```

### 3

Nel seguente esercizio è stato implementato il test del $\chi^2$ per verificare che è improbabile che il generatore di numeri casuali utilizzato non generi numeri veramente pseudo-casuali.
Sono stati generati $N=10^6$ numeri con distribuzione uniforme nell'intervallo $[0,1)$ e divisi in $N=100$ blocchi da $L = 10^4$ numeri ciascuno. Per ognuno di essi, sono stati contati il numero di punti $n_i$ ($i=1,2,3,..,100$) appertenti ad ognuno dei $100$ intervalli in cui è stato diviso $[0,1)$ per calcolare: $$\chi^2 = \sum_{i} \frac{\left( n_i - <n_i> \right)^2}{\sigma_{n_i}}$$  in cui <n_i> è il valore atteso del numero di punti per ognuno dei 100 intervalli ed $\sigma_{n_i}$ l'errore associato. Poichè la probabilità che si verifichino $n_i$ eventi su $n$, cioè che nell'intervallo i-esimo cadano esattamente $n_i$ punti, segue la distribuzione binomale la quale ha valore atteso il numero degli eventi per la probailità dell'evento $<n_i> = \frac{n}{M}$. L'errore $\sigma_{n_i}$ è la varianza della distribuzione $\sigma^2= n \frac{1}{M}(1- \frac{1}{M}) \approx  \frac{n}{M}$. Per cui si ottine:
$$\chi^2 = \sum_{i} \frac{\left( n_i - n/M \right)^2}{n/M}$$.


Il grafico seguente mostra che i $100$ valori del $\chi^2$ ottenuti sono centrati intorno al valore atteso, cioè $100$, ed hanno delle fluttauzioni. Non registrandosi alcuna anomalia, è lecito affermare che è estremamente improbabile che il generatore di numeri pseudi-casuali non sia considerabile casuale per ai fini dell'esercizio. 

```python
x,y= np.loadtxt("risultati01.1.3.out", usecols=(0,1), delimiter=' ', unpack='true')
plt.scatter(x,y)
plt.ylabel('$<\chi^2>$')
plt.title('$<\chi^2>$')
plt.grid(True)

plt.show()
```

## Esercizio 01.2

Sono stati calcolati $10^4$ volte gli $S_N$ così definiti: $$ S_N = \frac{1}{N}\sum_{i=1}^N x_i $$ con $N=1, 2, 10, 100$ e  $x_i$ una variabile casuale distribuita secondo le seguenti distribuzioni:
* uniforme: $P(x) = U \biggl( [0,1) \biggr)$
* esponenziale: $P(x)= \frac{1}{\lambda} e^{- \lambda x}$ con $\lambda=1$
* lorentziana: $P(x)= \frac{1}{\pi}\frac{\Gamma}{(x-\mu)^2+\Gamma^2}$ con $\mu=0$ e $\Gamma=1$.

Per campionare sia la distribuzione esponenziale che quella lorentziana è stato utilizzato il metodo dell'inversione della distribuzione di probabilità cumulata poichè richiede come prerequisito solo il generatore lineare congruenziale, la cui correttezza è stata verificata nell'esercizio precedente.
Sia $y$ una variable distributa uniformemente in $[0,1)$, allora è possibile generare un numero casuale distirbuito esponenzialmente $x$ tramite l'equazione $x = \frac{1}{\lambda} \log_{e}(1-y)$. Analogamente, se ne ricava uno distributo secondo una lorentziana con l'equazione: $x= \Gamma \tan [{\pi(y-\frac{1}{2}})]$.

Al crescere del numero delle variabili casuali sommate distribuite uniformemente e esponenzialmente, le $S_N$ tendono ad una guassiana con media $\mu = \langle S_N \rangle$ e deviazione standard $\sigma= \frac{ \sigma_N }{N}$  poichè gli $x_i$ utilizzati sono indipendenti, idienticamente distributiti ed estratti da due distribuzioni con varianza finita.
Gli $S_N$ ottenuti dagli  $x_i$ estratti dalla distribuzione lorentizana invece tendono ad una lorentziana poichè, pur essendo indipendenti ed identicamente distribuiti, sono estratti da una distribuzione lorentziana che ha varianza infinita e che ha come distribuzione limite una lorentziana.

Gli istogrammi normalizzati sottostanti mostrano che effettivamente i $S_N$ seguono le distribuzioni attese. Inoltre sono stati fittati i risultati degli $S_{100}$ per poter verificare graficamente l'aderenza dell'istogramma alla distribuzione attesa.

Lettura dei dadi da file: ogni colonna del file contiene 400000 risulati cioè le medie ottenute sommando 1,2,10,100 dadi.

```python
t=ascii.read("risultati01.2.out")
dado = np.asarray(t['col1'])
exp = np.asarray(t['col2'])
lor = np.asarray(t['col3'])
```

I seguono i quattro istogrammi relativi al lancio dei dadi con distribuzione uniforme normalizzati

```python
plt.hist(dado[0:10000],100, normed=True)
plt.xlabel('possibili esiti della media del lancio dei dadi')
plt.ylabel('numero di risultati')
plt.title('Un solo dado')
plt.show()

plt.hist(dado[10000:20000],100, normed=True)
plt.xlabel('possibili esiti della media del lancio dei dadi')
plt.ylabel('numero di risultati')
plt.title('Due dadi')
plt.show()

plt.hist(dado[20000:30000],100, normed=True)
plt.xlabel('possibili esiti della media del lancio dei dadi')
plt.ylabel('numero di risultati')
plt.title('Dieci dadi')
plt.show()

plt.hist(dado[30000:40000],30, normed=True)

mean,std=norm.fit(dado[30000:40000])
x = np.linspace(2.9, 4.1, 1000)
y = norm.pdf(x, mean, std)
plt.plot(x,y,label="fit con media 3.49 e deviazione standard 0.17")

plt.xlabel('possibili esiti della media del lancio dei dadi')
plt.ylabel('numero di risultati')
plt.title('Cento dadi')
plt.legend()
plt.show()
```

I seguono i quattro istogrammi relativi al lancio dei dadi con distribuzione esponenziale  con $\lambda = 1$ normalizzati

```python
plt.hist(exp[0:10000],100, color='green', normed=True)
plt.xlabel('possibili esiti della media del lancio dei dadi')
plt.ylabel('numero di risultati')
plt.title('Un solo dado')
plt.show()

plt.hist(exp[10000:20000],100, color='green', normed=True)
plt.xlabel('possibili esiti della media del lancio dei dadi')
plt.ylabel('numero di risultati')
plt.title('Due dadi')
plt.show()

plt.hist(exp[20000:30000],100, color='green', normed=True)
plt.xlabel('possibili esiti della media del lancio dei dadi')
plt.ylabel('numero di risultati')
plt.title('Dieci dadi')
plt.show()

plt.hist(exp[30000:40000],30, color='green', normed=True)

mean,std=norm.fit(exp[30000:40000])
x = np.linspace(0.5, 1.5, 1000)
y = norm.pdf(x, mean, std)
plt.plot(x,y,label="fit con media 0.999 e deviazione standard 0.098")

plt.xlabel('possibili esiti della media del lancio dei dadi')
plt.ylabel('numero di risultati')
plt.title('Cento dadi')
plt.legend()
plt.show()
```

I seguono i quattro istogrammi normalizzati relativi al lancio dei dadi con distribuzione lorentziana $\frac{1}{\pi}\frac{1}{x^2+1}$.

```python
plt.hist(lor[0:10000],bins=10000, color='orange', normed=True)
plt.xlim(-70,70)
plt.xlabel('possibili esiti della media del lancio dei dadi')
plt.ylabel('numero di risultati')
plt.title('Un solo dado')
plt.show()

plt.hist(lor[10000:20000],bins=10000, color='orange', normed=True)
plt.xlim(-70,70)
plt.xlabel('possibili esiti della media del lancio dei dadi')
plt.ylabel('numero di risultati')
plt.title('Due dadi')
plt.show()

plt.hist(lor[20000:30000],bins=10000, color='orange', normed=True)
plt.xlim(-70,70)
plt.xlabel('possibili esiti della media del lancio dei dadi')
plt.ylabel('numero di risultati')
plt.title('Dieci dadi')
plt.show()

plt.hist(lor[30000:40000],bins= 10000, color='orange', normed=True)
plt.xlim(-70,70)

mean,std=cauchy.fit(lor[30000:40000])
x = np.linspace(-20, 20)
y = cauchy.pdf(x, mean, std)

plt.plot(x,y,label="fit con media -0.01")
plt.xlabel('possibili esiti della media del lancio dei dadi')
plt.ylabel('numero di risultati')
plt.title('Cento dadi')
plt.legend()
plt.show()
```

## Esercizio 01.3

E' stato realizzato $M=10^4$ volte l'esperimento di Buffon per ricavare $\pi$: $$\pi = \frac{2L}{Pd} = \lim_{N_{thr} \to \infty}\frac{2LN_{thr}}{N_{hit}d}$$
con $N_{thr} = 50000 $  lunghezza dell'ago $L = 4$ e distanza tra le linee $d=6$.

Per fare ciò, è stata generata la posizione verticale del centro dell'ago $y$ uniformemente in $[-\frac{d}{2}, \frac{d}{2}]$ ( l'unica coordinata necessaria poichè il problema è invariante per traslazioni lungo l'asse orizzionatale) ed è stato generato il seno dell'angolo di rotazione rispetto alla verticale per ricavare la coordinate $y_1$ ed $y_2$ degli estremi dell'ago $y_{1/2}= y \pm \frac{L}{2} \sin(\theta) $. Per ricavare $ \sin\theta$ è stata generata una coppia di numeri casuali $(x,y)$ distribuiti uniformemente in $[-1,1)$ e, se la distanza dall'origine $\sqrt{x^2+y^2}$ è inferiore ad $1$, si ottiene $ \sin\theta= \frac{y}{\sqrt{x^2+y^2}}$. In tal modo si evita di utilizzare la conoscenza di $\pi$ per generare angolo uniformemente in $[0, 2\pi]$.

Ogni volta che l'ago interseca la retta orizzionatale di equazione $y=0$, cioè se viene verificata la condizione $( y_1 \ge 0 \land  \;  y_2 \le 0) \lor (y_1 \le \land \; y_2 \ge 0)$, il contatore dei successi $N_{hit}$ viene incrementato di $1$.

Utilizzando il metodo del data blocking con $N = 100$ blocchi da $L = 100$ dati ciascuno, ognuno dei quali è ottenuto con $M$ realizzazioni dell'esperimento, è stato realizzato il grafico sottostante da cui si evince che l'esperimento converge al valore $\pi= 3.1415$ entro l'errore associato.

```python
x, y, error = np.loadtxt("risultati01.3.out", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,y - 3.1415 ,yerr=error)
plt.xlabel('N')
plt.ylabel("$< \pi > $ - 3.1415")
plt.title('$< \pi > $')
plt.grid(True)]

plt.show()
```
