# 11.1

Per studiare il funzionamento delle rete neurale sono stati variati i seguenti parametri: il numero di epoche $N_{\mathrm{epochs}}$, il numero dei dati di di trian $N_{\mathrm{train}}$ e il rumore di dati $\sigma$ di input.

Il grafico *model loss* sottostante mostra che inizialmente la loss function di train e test non coincidono, a $35$ epoche coincidono e oltre tale valore tendono a zero e si discostano poco l'una dall'altra.
Inoltre aumentando il numero di dati di train il fit migliora ma al crescere dell'errore $\sigma$ peggiora.

```python
# target parameters of f(x) = m*x + b
m = 2 # slope
b = 1 # intersect

np.random.seed(0)
x_train = np.random.uniform(-1, 1, 200)
x_valid = np.random.uniform(-1, 1, 20)
x_valid.sort()
y_target = m * x_valid + b # ideal (target) linear function

sigma = 0.5# noise standard deviation, for the moment it is absent
y_train = np.random.normal(m * x_train + b, sigma) # actual measures from which we want to guess regression parameters
y_valid = np.random.normal(m * x_valid + b, sigma)

plt.plot(x_valid, y_target)
plt.scatter(x_valid, y_valid, color='r')
plt.grid(True); plt.show()
```

```python
model = tf.keras.Sequential()
model.add(Dense(1, input_shape=(1,)))

# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])

# get a summary of our composed model
model.summary()
```

```python
# fit the model using training dataset
# over 10 epochs of 32 batch size each
# report training progress against validation data
history = model.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=70,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid))
```

```python
# evaluate model
score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=1)

# print performance
print('Valid')
print('Test loss:', score[0])
print('Test accuracy:', score[1])
print()

# evaluate model with the exact curve
score = model.evaluate(x_valid, y_target, batch_size=32, verbose=1)

# print performance
print('Target')
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# look into training history
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()

x_predicted = np.random.uniform(-1, 1, 100)
y_predicted = model.predict(x_predicted)
plt.scatter(x_predicted, y_predicted,color='r')
plt.plot(x_valid, y_target)
plt.grid(True)
plt.show()
```

```python
# generate training inputs
np.random.seed(0)
x_train = np.random.uniform(-1, 1, 2000)
x_valid = np.random.uniform(-1, 1, 200)
x_valid.sort()

y_target = 4 -3* x_valid -2*(x_valid**2) + 3*(x_valid**3) # ideal (target) linear function

sigma = 0.0# noise standard deviation, for the moment it is absent
y_train = np.random.normal(4 -3* x_train -2*(x_train**2) + 3*(x_train**3), sigma) # actual measures from which we want to guess regression parameters
y_valid = np.random.normal(4 -3* x_valid -2*(x_valid**2) + 3*(x_valid**3), sigma)

plt.plot(x_valid, y_target)
plt.scatter(x_valid, y_valid, color='r')
plt.grid(True)
plt.show()
```

```python
model = Sequential()

model.add(Dense(1, input_shape=(1,)))

model.add(Dense(60, activation='tanh'))
model.add(Dense(40, activation='relu'))
model.add(Dense(20, activation='softsign'))

model.add(Dense(1, ))

model.compile(optimizer='sgd',  loss='mse', metrics=['mse'])
model.summary()
```

```python
bsize = 32
# fit the model using training dataset
# over 50 epochs of 32 batch size each
# report training progress against validation data
history = model.fit(x=x_train, y=y_train, 
          batch_size=bsize, epochs=40, 
          validation_data=(x_valid, y_valid))
```

```python
# evaluate model
score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=1)

# print performance
print('Trainig e test')
print('Test loss:', score[0])
print('Test accuracy:', score[1])

# evaluate model with the exact curve
score = model.evaluate(x_valid, y_target, batch_size=32, verbose=1)

# print performance
print('Performance')
print('Test loss:', score[0])
print('Test accuracy:', score[1])


plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()


#teorico
nuovo_estremo=1.2
x = np.random.uniform(-nuovo_estremo, nuovo_estremo, 100)
x.sort()
y = 4 -3* x -2*(x**2) + 3*(x**3)
plt.plot(x, y, alpha=0.6, label='valore esatto del polinomio')
#predetti
x_predicted = np.random.uniform(-1, 1, 300)
y_predicted = model.predict(x_predicted)
plt.scatter(x_predicted, y_predicted,color='r', label='predetti in intervallo di test')

#predetti in intervallo esteso
x_predicted = np.random.uniform(-nuovo_estremo, -1, 100)
y_predicted = model.predict(x_predicted)
plt.scatter(x_predicted, y_predicted,color='g')
x_predicted = np.random.uniform(1, nuovo_estremo, 100)
y_predicted = model.predict(x_predicted)
plt.scatter(x_predicted, y_predicted,color='g', label='predetti fuori da intervallo di test')
plt.title('Previsioni')

plt.grid(True)
plt.legend()
#plt.savefig('Risultati/7.png')
plt.show()
```

# 11.3

La rete neurale costruita per provare a fittare la funzione trigonometrica 
$$ f(x,y) = \sin(x^2+y^2) \quad \text{per} \quad x \in [-3/2,3/2] \; \text{e} \;  y \in [-3/2,3/2]\, $$
presenta la seguente architettura:

  $l$| $N_l$ | $A_l$
-----| ----- | -----
1    |   2   | -
2    |   50  | tanh
3    |   50  | relu
4    |   1   | -

La rete è stata addestrata con $N_{\mathrm{train}}=10^4$ dati generati casualmente utilizzando come ottimizzatore lo stochastic gradient descent, come loss function il mean square error e come metodo per la valutazione, ancora un volta, il mean square error. Dopo un'addrestamento di $N_{\mathrm{epoch}}=60$, la performance della rete è stata valutata con $N_{\mathrm{test}}=10^3$ ed è riportata nel grafico *Model loss*. S osserva che la rete non ha overfitting ed ha ottime prestazioni. Inoltre è stato realizzato un grafico per confrontare i valori esatti della funzione con quelli predetti, anche al di fuori dell'intervallo di test, generando i dati casualmente $x \in [-2,2] \; \text{e} \;  y \in [-2,2]\,$. Sembra esservi buona aderenza tra la funzione e i dati prendetti anche al di fuori dell'intervallo dei dati di train.

```python
def sin(x):
    return np.sin(x[:,0]**2 + x[:,1]**2)
```

```python
# generate training inputs
#np.random.seed(0)

n_dati=10000
n_valid=1000
estremo=1.5

#dati per il train matrice degli x_ sono gli imput della funzione (x,y) e il vettore degli y_ è il valore della funzione in (x,y)
x_train=np.zeros((n_dati,2))
x_train[:,0] = np.random.uniform(-estremo, estremo, n_dati)
x_train[:,1] = np.random.uniform(-estremo, estremo, n_dati)
y_train = sin(x_train)


#dati per il validazione matrice degli x_ sono gli imput della funzione (x,y) e il vettore degli y_ è il valore della funzione in (x,y)
x_valid=np.zeros((n_valid,2))
x_valid[:,0] = np.random.uniform(-estremo, estremo, n_valid)
x_valid[:,1] = np.random.uniform(-estremo, estremo, n_valid)
y_valid = sin(x_valid) 

y_target = sin(x_valid) #dati esatti da utilizzare per grafici

#possibilità di aggiungere del white noise ai dati
#sigma = 0.0
#y_train = np.random.normal(y_train, sigma)
#y_valid = np.random.normal(y_valid, sigma)


#grafico 3D dei dati di validazione
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(x_train[:,0], x_train[:,1], y_train, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 30)
plt.show()
```

```python
# morfologia della rete neurale
model = tf.keras.Sequential()
model.add(Dense(2, input_shape=(2,)))
model.add(Dense(50,activation='tanh'))
model.add(Dense(50,activation='relu'))
model.add(Dense(1))

model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
model.summary()
```

```python
history = model.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=60,
          shuffle=True,
          validation_data=(x_valid, y_valid))
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()
```

```python
# evaluate model with the exact curve
score = model.evaluate(x_valid, y_target, batch_size=32, verbose=1)

estremo=2
n_predicted = 1000
x_predicted = np.zeros((n_predicted, 2))
y_predicted = np.zeros((n_predicted))
x_predicted[:,0] = np.random.uniform(-estremo, estremo, n_predicted)
x_predicted[:,1] = np.random.uniform(-estremo, estremo, n_predicted)
y_predicted= model.predict(x_predicted)


fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(x_valid[:,0], x_valid[:,1], y_target, c='red', marker='.', label='target')
ax.scatter(x_predicted[:,0], x_predicted[:,1],y_predicted, c='blue', marker='.',label = 'predetti')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x,y)')
plt.legend()
plt.title('Confronto predizioni e modello')
plt.show()
```
