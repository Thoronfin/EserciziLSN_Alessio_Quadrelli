# 10.1

Per risolvere il problema del commesso viaggiatore tra $30$ città disposte o in un quadrato di lato $l=1$ o lungo una circonferenza di raggio $r=1$ è stato implementato il simulated annealing come segue:

* **Inizilazzazione:** come per l'esercizio $9.1$ vengono iniziate casualmente le posizioni delle $i$ città (nel quadrato o lungo la circonferenza) $(x_i,y_i)$ e, permutandole casualmente, viene generata la configurazione inizale che può essere formata da un numero variabile di percorsi. In tutti i gli esercizi della lezione è stata utilizzata una popolazione di un solo percorso.

*  **Metropolis:** una nuova possibile configurazione $\vec{x}^{'}$ è generata con $N_{mutazioni}$ con probabilità $p$ della configurazione attuale  $\vec{x}$, cioè con $N_{mutazioni}$ permutazioni casuali di una coppia di città. La configurazione proposta viene accettata sicuramente se $L^{(2)}(\vec{x}^{'}) < L^{(2)}(\vec{x})$, altrimenti viene accettata con probabilità $P = e^{\beta \left(L^{(2)}(\vec{x}^{'}) -L^{(2)}(\vec{x})\right)}$,
dove $L^{(2)}(\vec{x}) = \sum_{i} (x_{i+1} - x_i)^2 + (y_{i+1} - y_i)^2$.

E' stata utilizzata una sola mutazione ($N_{mutazioni} = 1$) con probabilità $p=1$ così che sia a bassa temperatura è possibile passare da una configurazione ad una nuova (cosa che sarebbe estremamente improbabile con più mutazioni) sia vi è una certa variabilità che permette di esplorare lo spazio delle configurazioni.

L'algoritmo è stato utilizzato variando $\beta$, da $\beta_{iniziale}$ a $\beta_{finale}$, di $\Delta \beta$ con i seguenti parametri per i due casi:

 |Quadrato |Cerchio
--|---
$\beta_{iniziale}$|3|0
$\beta_{finale}$|40|40
$\Delta \beta$|0.009|0.006
Campionamenti per $\beta$ |2000|2000
$N_{mutazioni}$|1|1
$p$|1|1

I grafici sottostanti riportano la minima distanza, calcolata tra i $2000$ campionamenti in funzione di $\beta$, e il miglior percorso del commesso viaggiatore in entrambi i casi.

Si osserva che il miglior percorso nel caso di città disposte nel quadrato ha lunghezza $0.937$ che è minore rispetto a quella ottenuta nell'esercizio $9.1$ e nel $10.2$. Variando i parametri non è stato possible migliorare il risultato ma solo perggiorarlo. Invece il miglior percorso lungo la circonferenza è esattamente uguale a quello dell'esercizio $9.1$ e $10.2$.

```python
############Quadrato##############################
plt.figure(figsize=(17,12))
plt.suptitle('Quadrato di lato 1') 

plt.subplot(2,2,1)
y = np.loadtxt("10.1/Distanze_quadrato.dat", usecols=(0), delimiter=' ', unpack='true')
x=np.arange(3,40,0.009)
print('minima distanza quadrato: ', y[len(y)-1])
plt.plot(x,y)
plt.xlabel(r'$\beta$')
plt.ylabel('Distanza percorsa')
plt.title('MIglior distanza percorsa')
plt.grid(True)

plt.subplot(2,2,2)
n,x,y= np.loadtxt("10.1/BestPath_quadrato.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.plot(x,y, marker='.', label='città', color='blue', linestyle='--')
plt.plot(x[0],y[0], marker='.', markersize=20, label='inizio e fine percorso', color='red')
plt.plot(x[1],y[1], marker='.', markersize=20, label='prima tappa', color='green')
plt.plot(x[len(x)-2],y[len(y)-2], marker='.', markersize=20, label='penultima tappa', color='orange')
plt.title('MIglior percorso del commesso viaggiatore')
plt.legend()
plt.grid(True)
plt.show()


############Quadrato##############################
plt.figure(figsize=(17,12))
plt.suptitle('Cerchio di raggio 1') 

plt.subplot(2,2,1)
y = np.loadtxt("10.1/Distanze_cerchio.dat", usecols=(0), delimiter=' ', unpack='true')
print('minima distanza cerchio: ', y[len(y)-1])
x=np.arange(0,40,0.006)
plt.plot(x,y)
plt.xlabel(r'$\beta$')
plt.ylabel(r'Distanza percorsa')
plt.title('Minima distanza percorsa')
plt.grid(True)

plt.subplot(2,2,2)
n,x,y= np.loadtxt("10.1/BestPath_cerchio.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.plot(x,y, marker='.', label='città', color='blue', linestyle='--')
plt.plot(x[0],y[0], marker='.', markersize=20, label='inizio e fine percorso', color='red')
plt.plot(x[1],y[1], marker='.', markersize=20, label='prima tappa', color='green')
plt.plot(x[len(x)-2],y[len(y)-2], marker='.', markersize=20, label='penultima tappa', color='orange')
plt.title('Miglior percorso del commesso viaggiatore')
plt.legend()
plt.grid(True)
plt.show()
```

# 10.2

Per eseguire il random search su $N$ nodi con il simulated annealing è stato modificato il codice precedente imponendo che vi sia una sola mutazione e passando al costruttore della classe un parametro che servisse per avere permutazioni diverse delle città in modo che ogni nodo abbia una configurazione iniziale diversa ma che in tutti i nodi le città abbiano la stessa posizione poichè in questo caso il seme è letto da un file uguale per ognuno.

A differenza dell'esercizio $10.1$, dopo i campionamenti a $\beta$ fissato, tutti i nodi calcolano la miglior distanza raggiunta da loro fino a quel momento e la inviano al nodo $0$, il quale seleziona quella migliore e la stampa su file. Così sono stati realizzati i grafici della miglior distanza percorsa. All'ultimo valore di $\beta$, tutti i nodi inviano la miglior distanza percorsa al nodo $0$, il quale determina il valore minimo e comunica al nodo che lo ha raggiunto di stampare su file il percorso migliore (riportato nei grafici sottostanti). La comunicazione tra i nodi per ogni valore di $\beta$ peggiora le prestazioni del programma nettamente. Per migliorare l'efficienza si può fare lo scambio delle informazoni solo a $\beta_{finale}$ e si otterrebe la medesima miglior distanza e miglior percorso senza però avere informazioni sull'evoluzione del proccesso al variare di $\beta$.

L'algoritmo è stato utilizzato variando $\beta$, da $\beta_{iniziale}$ a $\beta_{finale}$, di $\Delta \beta$ con i seguenti parametri per i due casi:

 |Quadrato |Cerchio
--|---|---
$\beta_{iniziale}$|3|0
$\beta_{finale}$|40|20
$\Delta \beta$|0.009|0.006
Campionamenti per $\beta$ |2000|2000
$p$|1|1
$N$|4|4

Si osserva che nel caso del quadrato è stata ottenuta la minima distanza rispetto alla lezione $9.1$ e all'esercizio $10.1$, mentre per il cerchio è identica a quelle già ottenute.

```python
############Quadrato##############################
plt.figure(figsize=(17,12))
plt.suptitle('Quadrato di lato 1') 

plt.subplot(2,2,1)
y = np.loadtxt("10.2/Distanze_quadrato.dat", usecols=(0), delimiter=' ', unpack='true')
x=np.arange(3,40,0.009)
print('minima distanza quadrato: ', y[len(y)-1])
plt.plot(x,y)
plt.xlabel(r'$\beta$')
plt.ylabel('Distanza percorsa')
plt.title('MIglior distanza percorsa')
plt.grid(True)

plt.subplot(2,2,2)
n,x,y= np.loadtxt("10.2/BestPath_quadrato.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.plot(x,y, marker='.', label='città', color='blue', linestyle='--')
plt.plot(x[0],y[0], marker='.', markersize=20, label='inizio e fine percorso', color='red')
plt.plot(x[1],y[1], marker='.', markersize=20, label='prima tappa', color='green')
plt.plot(x[len(x)-2],y[len(y)-2], marker='.', markersize=20, label='penultima tappa', color='orange')
plt.title('MIglior percorso del commesso viaggiatore')
plt.legend()
plt.grid(True)
plt.show()


############Quadrato##############################
plt.figure(figsize=(17,12))
plt.suptitle('Cerchio di raggio 1') 

plt.subplot(2,2,1)
y = np.loadtxt("10.2/Distanze_cerchio.dat", usecols=(0), delimiter=' ', unpack='true')
print('minima distanza cerchio: ', y[len(y)-1])
x=np.arange(0,20,0.006)
plt.plot(x,y)
plt.xlabel(r'$\beta$')
plt.ylabel(r'Distanza percorsa')
plt.title('Minima distanza percorsa')
plt.grid(True)

plt.subplot(2,2,2)
n,x,y= np.loadtxt("10.2/BestPath_cerchio.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.plot(x,y, marker='.', label='città', color='blue', linestyle='--')
plt.plot(x[0],y[0], marker='.', markersize=20, label='inizio e fine percorso', color='red')
plt.plot(x[1],y[1], marker='.', markersize=20, label='prima tappa', color='green')
plt.plot(x[len(x)-2],y[len(y)-2], marker='.', markersize=20, label='penultima tappa', color='orange')
plt.title('Miglior percorso del commesso viaggiatore')
plt.legend()
plt.grid(True)
plt.show()
```
