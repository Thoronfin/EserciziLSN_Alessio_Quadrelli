## 3.1

Per il simulare l'evoluzione del prezzo degli assets nel tempo - $S(t)$ - da $t=0$ a $t=T=1$, con volatilità $\sigma$ e drift $\mu=r$ (in questo caso coincide con il tasso d'interesse) costanti, è stato utilizzato il moto browinano geometrico come nella trattazione di Black e Scholes. Il moto browniano geometrico, che è un processo continuo nel tempo, può essere campionato in istanti discreti $t_i$ senza commettere errori di discretizzazione poichè l'incremento dipende solo dalla differenza tra i tempi dei campionamenti $t_{i+1}-t_i$ :  $$S(t_{i+1}) = S(t_i) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)
    +\sigma Z_{i+1}\sqrt{t_{i+1}-t_i}\right] \quad \text{con} \quad Z_i \sim \mathcal{N}(0,1)\, .$$
    
Poiché è indipendente dalla discretizzazione, il prezzo degli assets è stato campionato dividendo l'intervallo di tempi $[0,1]$ in $100$ intervalli uguali e non dividendolo affatto.
   

I prezzi da assegnare alle call e alle put options sono i seguenti:

* $C\left[ S(T),T \right] = e^{-rT}\max\left[0,S(T)-K\right]$
* $P\left[ S(T),T \right] = e^{-rT}\max\left[0,K-S(T)\right]$

dove $K$ è lo strike price e il termine  $e^{-rT}$ tiene conto del fatto che, depositando una quantità di soldi $S(t=0)$ in banca con un tasso di interesse $r$, dopo un tempo $t=T$, questi sarebbero diventati $S(t=T)= e^{rT} S(t=0)$ senza rischi.

Nei grafici sottostanti sono riportati i prezzi da assegnare alla call-option e alla put-option europee al tempo $T=1$ (tenendo conto del tasso di interesse senza rischio $r=0.1$, della volatilità $\sigma=0.25$, dello strike price $K=100$ e di $S(t=0)=100$. Questi sono stati calcolati mediante il metodo del data blocking con $N = 100$ blocchi da $L = 100$ dati ciascuno. I risultati ottenuti sono riportati nei grafici sottostanti e sono stati confrontati con i valori esatti ottenuti dalla risoluzione analitica dell'equazione di Black-Scholes: si osserva che convergono, entro l'errore associato, al valore esatto.

```python
plt.figure(figsize=(17,10))

plt.subplot(2,2,1)

x, y, error = np.loadtxt("risultati03.call_un_passo.out", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,y-14.975, yerr=error, label='Un passo')
x, y, error = np.loadtxt("risultati03.call_passi.out", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,y-14.975, yerr=error, label='100 passi')


plt.xlabel('N')
plt.ylabel('Call-option price -14.975')
plt.title(r'Call-option price')


plt.legend()
plt.grid(True)

plt.subplot(2,2,2)
x, y, error = np.loadtxt("risultati03.put_un_passo.out", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,y-5.4595, yerr=error, label='Un passo')
x, y, error = np.loadtxt("risultati03.put_passi.out", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,y-5.4595, yerr=error, label='100 passi')

plt.xlabel('N')
plt.ylabel('Put-option price -5.4595')
plt.legend()
plt.grid(True)
plt.title(r'Put-option price ')


plt.show()
```
